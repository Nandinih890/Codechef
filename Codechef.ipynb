```python
n = int(input())
result = 0

for _ in range(n):
    result ^= int(input())

print(result)
```

```python
def can_place(stalls, cows, distance):
    count = 1
    last = stalls[0]

    for i in range(1, len(stalls)):
        if stalls[i] - last >= distance:
            count += 1
            last = stalls[i]

        if count >= cows:
            return True

    return False

n, cows = map(int, input().split())
stalls = [int(input()) for _ in range(n)]

stalls.sort()

low = 1
high = stalls[-1] - stalls[0]
answer = 0

while low <= high:
    mid = (low + high) // 2

    if can_place(stalls, cows, mid):
        answer = mid
        low = mid + 1
    else:
        high = mid - 1

print(answer)
```

```python
withdraw, balance = input().split()

withdraw = int(withdraw)
balance = float(balance)

if withdraw % 5 == 0 and withdraw + 0.50 <= balance:
    balance -= (withdraw + 0.50)

print(f"{balance:.2f}")
```
